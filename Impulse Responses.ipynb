```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

# Impulse Responses

```python
irf_list = list(pd.read_csv('Data/ir_data/irf_df.csv').iloc[:,1:].columns)
print(irf_list)
irf = pd.read_csv('Data/ir_data/irf_df.csv').iloc[:,1:]
lower = pd.read_csv('Data/ir_data/lower_df.csv').iloc[:,1:]
upper = pd.read_csv('Data/ir_data/upper_df.csv').iloc[:,1:]

df_dict = {}
for n in irf_list:
    df = pd.DataFrame({'irf': irf[n], 'lower': lower[n], 'upper': upper[n]})
    df_dict[n] = df
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))
plt.subplots_adjust(wspace=0.2, hspace=0.3, top=0.8)
fig.suptitle('Structural Impulse Responses', fontsize=13)

axes[0].plot(df_dict['candidate_ol2candidate'].irf, lw = 1.8)
axes[0].plot(df_dict['candidate_ol2candidate'].lower, lw = 1, linestyle='-.', color='black')
axes[0].plot(df_dict['candidate_ol2candidate'].upper, lw = 1, linestyle='-.', color='black')
axes[0].axhline(y=0, color='black', lw=1.8)
axes[0].set_title('candidate_ol -> candidate')

axes[1].plot(df_dict['candidate2candidate_ol'].irf, lw = 1.8)
axes[1].plot(df_dict['candidate2candidate_ol'].lower, lw = 1, linestyle='-.', color='black')
axes[1].plot(df_dict['candidate2candidate_ol'].upper, lw = 1, linestyle='-.', color='black')
axes[1].axhline(y=0, color='black', lw=1.8)
axes[1].set_title('candidate -> candidate_ol')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))
plt.subplots_adjust(wspace=0.2, hspace=0.3, top=0.8)
fig.suptitle('Structural Impulse Responses', fontsize=13)

axes[0].plot(df_dict['party_ol2party'].irf, lw = 1.8)
axes[0].plot(df_dict['party_ol2party'].lower, lw = 1, linestyle='-.', color='black')
axes[0].plot(df_dict['party_ol2party'].upper, lw = 1, linestyle='-.', color='black')
axes[0].axhline(y=0, color='black', lw=1.8)
axes[0].set_title('party_ol -> party')

axes[1].plot(df_dict['party2party_ol'].irf, lw = 1.8)
axes[1].plot(df_dict['party2party_ol'].lower, lw = 1, linestyle='-.', color='black')
axes[1].plot(df_dict['party2party_ol'].upper, lw = 1, linestyle='-.', color='black')
axes[1].axhline(y=0, color='black', lw=1.8)
axes[1].set_title('party -> party_ol')
```
